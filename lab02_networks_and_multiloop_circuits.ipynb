1. Calculate the equivalent resistance $R_{\text{parallel}}$ for the parallel resistors $R_{\text{2}}$ and $R_{\text{3}}$, and the total equivalent resistance $R_{\text{equiv}}$ of the circuit

```python
# Given values
print("=== GIVEN VALUES ===")
print(f"R1 = {R1} Ω")
print(f"R2 = {R2} Ω") 
print(f"R3 = {R3} Ω")
print(f"E1 = {E1_theor} V")

# Step 1: Calculate equivalent resistance for parallel resistors R2 and R3
# For parallel resistors: 1/R_parallel = 1/R2 + 1/R3
# Or: R_parallel = (R2 × R3) / (R2 + R3)

print("\n=== PARALLEL RESISTANCE CALCULATION ===")
print("For R2 and R3 in parallel:")
print("Method 1 - Product over Sum formula:")
R_parallel = (R2 * R3) / (R2 + R3)
print(f"R_parallel = (R2 × R3) / (R2 + R3)")
print(f"R_parallel = ({R2} × {R3}) / ({R2} + {R3})")
print(f"R_parallel = {R2 * R3:,} / {R2 + R3}")
print(f"R_parallel = {R_parallel:.2f} Ω")

print("\nMethod 2 - Reciprocal formula (verification):")
reciprocal_sum = (1/R2) + (1/R3)
R_parallel_alt = 1 / reciprocal_sum
print(f"1/R_parallel = 1/R2 + 1/R3")
print(f"1/R_parallel = 1/{R2} + 1/{R3}")
print(f"1/R_parallel = {1/R2:.8f} + {1/R3:.8f}")
print(f"1/R_parallel = {reciprocal_sum:.8f}")
print(f"R_parallel = {R_parallel_alt:.2f} Ω")

# Step 2: Calculate total equivalent resistance
# R1 is in series with the parallel combination of R2 and R3
# For series: R_equiv = R1 + R_parallel

print("\n=== TOTAL EQUIVALENT RESISTANCE ===")
R_equiv = R1 + R_parallel
print("R1 is in series with the parallel combination (R2 || R3):")
print(f"R_equiv = R1 + R_parallel")
print(f"R_equiv = {R1} + {R_parallel:.2f}")
print(f"R_equiv = {R_equiv:.2f} Ω")

# Summary
print(f"\n=== SUMMARY ===")
print(f"Parallel resistance (R2 || R3): {R_parallel:.2f} Ω")
print(f"Total equivalent resistance: {R_equiv:.2f} Ω")

# Store values for later use
print(f"\n=== STORED VALUES ===")
print(f"R_parallel = {R_parallel:.2f} Ω")
print(f"R_equiv = {R_equiv:.2f} Ω")
```

2. Calculate the theoretical value $i_{\text{1}}$, the total current in the circuit, and the voltage drop across $R_{\text{1}}$

```python
# Step 1: Calculate the total current i1 using Ohm's Law
# In a series-parallel circuit, i1 is the total current flowing from the source
# Using Ohm's Law: I = V/R

print("=== TOTAL CURRENT CALCULATION ===")
print("Using Ohm's Law: I = V/R")
print(f"Applied voltage: E1 = {E1_theor} V")
print(f"Total equivalent resistance: R_equiv = {R_equiv:.2f} Ω")

i1_theoretical = E1_theor / R_equiv
print(f"\nTotal current calculation:")
print(f"i1 = E1 / R_equiv")
print(f"i1 = {E1_theor} V / {R_equiv:.2f} Ω")
print(f"i1 = {i1_theoretical:.6f} A")
print(f"i1 = {i1_theoretical * 1000:.3f} mA")

# Step 2: Calculate voltage drop across R1
# Since R1 is in series with the parallel combination, the total current i1 flows through R1
# Using Ohm's Law: V = I × R

print(f"\n=== VOLTAGE DROP ACROSS R1 ===")
print("Since R1 is in series, the total current i1 flows through it:")
print(f"Using Ohm's Law: V = I × R")

V1_theoretical = i1_theoretical * R1
print(f"\nVoltage drop across R1:")
print(f"V1 = i1 × R1")
print(f"V1 = {i1_theoretical:.6f} A × {R1} Ω")
print(f"V1 = {V1_theoretical:.6f} V")
print(f"V1 = {V1_theoretical:.3f} V")

# Verification: Check that V1 + V_parallel = E1
V_parallel_check = E1_theor - V1_theoretical
print(f"\n=== VERIFICATION ===")
print("Kirchhoff's Voltage Law: V1 + V_parallel = E1")
print(f"V_parallel = E1 - V1 = {E1_theor} - {V1_theoretical:.3f} = {V_parallel_check:.3f} V")
print(f"This voltage will appear across the parallel combination of R2 and R3")

# Summary
print(f"\n=== SUMMARY FOR QUESTION 2 ===")
print(f"Total current in circuit: i1 = {i1_theoretical:.6f} A = {i1_theoretical * 1000:.3f} mA")
print(f"Voltage drop across R1: V1 = {V1_theoretical:.3f} V")
print(f"Voltage across parallel network: V_parallel = {V_parallel_check:.3f} V")

# Store values for later calculations
print(f"\n=== STORED VALUES ===")
print(f"i1_theoretical = {i1_theoretical:.6f} A")
print(f"V1_theoretical = {V1_theoretical:.3f} V")
```

```python
# Step 1: Calculate voltage drop across the parallel network
# From previous calculation: V_parallel = E1 - V1
# We already calculated this in the previous step

V_parallel = E1_theor - V1_theoretical

print("=== VOLTAGE DROP ACROSS PARALLEL NETWORK ===")
print("Using Kirchhoff's Voltage Law:")
print(f"V_parallel = E1 - V1")
print(f"V_parallel = {E1_theor} V - {V1_theoretical:.3f} V")
print(f"V_parallel = {V_parallel:.3f} V")
print()
print("This voltage appears across both R2 and R3 since they are in parallel")

# Step 2: Calculate current through R2
# In parallel circuits, voltage across each resistor is the same (V_parallel)
# Using Ohm's Law: I = V/R

print(f"\n=== CURRENT THROUGH R2 ===")
print("Using Ohm's Law: I = V/R")
print(f"Voltage across R2: V2 = V_parallel = {V_parallel:.3f} V")
print(f"Resistance R2 = {R2} Ω")

i2_theoretical = V_parallel / R2
print(f"\nCurrent through R2:")
print(f"i2 = V2 / R2")
print(f"i2 = {V_parallel:.3f} V / {R2} Ω")
print(f"i2 = {i2_theoretical:.6f} A")
print(f"i2 = {i2_theoretical * 1000:.3f} mA")

# Step 3: Calculate current through R3
print(f"\n=== CURRENT THROUGH R3 ===")
print("Using Ohm's Law: I = V/R")
print(f"Voltage across R3: V3 = V_parallel = {V_parallel:.3f} V")
print(f"Resistance R3 = {R3} Ω")

i3_theoretical = V_parallel / R3
print(f"\nCurrent through R3:")
print(f"i3 = V3 / R3")
print(f"i3 = {V_parallel:.3f} V / {R3} Ω")
print(f"i3 = {i3_theoretical:.6f} A")
print(f"i3 = {i3_theoretical * 1000:.3f} mA")

# Step 4: Verify Kirchhoff's Current Law
# At the junction where the current splits: i1 = i2 + i3
current_sum = i2_theoretical + i3_theoretical

print(f"\n=== VERIFICATION - KIRCHHOFF'S CURRENT LAW ===")
print("At the junction: i1 = i2 + i3")
print(f"i2 + i3 = {i2_theoretical:.6f} + {i3_theoretical:.6f}")
print(f"i2 + i3 = {current_sum:.6f} A")
print(f"i1 (from previous calculation) = {i1_theoretical:.6f} A")
print(f"Difference: {abs(i1_theoretical - current_sum):.8f} A")

if abs(i1_theoretical - current_sum) < 1e-8:
    print("✓ Kirchhoff's Current Law verified: i1 = i2 + i3")
else:
    print("✗ Error in calculations - check your work")

# Step 5: Verify current relationship for parallel resistors
# Since R3 > R2, we expect i2 > i3
current_ratio = i2_theoretical / i3_theoretical
resistance_ratio = R3 / R2

print(f"\n=== CURRENT-RESISTANCE RELATIONSHIP ===")
print(f"Since R3 ({R3} Ω) > R2 ({R2} Ω), we expect i2 > i3")
print(f"Current ratio i2/i3 = {current_ratio:.2f}")
print(f"Resistance ratio R3/R2 = {resistance_ratio:.2f}")
print(f"These should be equal for parallel resistors: ✓" if abs(current_ratio - resistance_ratio) < 0.01 else "These should be equal: ✗")

# Summary
print(f"\n=== SUMMARY FOR QUESTION 3 ===")
print(f"Voltage across parallel network: V_parallel = {V_parallel:.3f} V")
print(f"Current through R2: i2 = {i2_theoretical:.6f} A = {i2_theoretical * 1000:.3f} mA")
print(f"Current through R3: i3 = {i3_theoretical:.6f} A = {i3_theoretical * 1000:.3f} mA")
print(f"Verification: i1 = i2 + i3 = {current_sum:.6f} A ✓")

# Store values for later calculations
print(f"\n=== STORED VALUES ===")
print(f"V_parallel = {V_parallel:.3f} V")
print(f"i2_theoretical = {i2_theoretical:.6f} A")
print(f"i3_theoretical = {i3_theoretical:.6f} A")

# Create summary table
print(f"\n=== COMPLETE CIRCUIT ANALYSIS TABLE ===")
print(f"{'Component':<12} {'Resistance (Ω)':<15} {'Voltage (V)':<12} {'Current (mA)':<12}")
print("-" * 55)
print(f"{'R1':<12} {R1:<15} {V1_theoretical:<12.3f} {i1_theoretical*1000:<12.3f}")
print(f"{'R2':<12} {R2:<15} {V_parallel:<12.3f} {i2_theoretical*1000:<12.3f}")
print(f"{'R3':<12} {R3:<15} {V_parallel:<12.3f} {i3_theoretical*1000:<12.3f}")
print(f"{'Total':<12} {R_equiv:<15.0f} {E1_theor:<12.3f} {i1_theoretical*1000:<12.3f}")
```

<h4 style="text-align: center">The Multiloop Circuit </h4>

1. Set up the equations that govern the multiloop circuit

```python
# MULTILOOP CIRCUIT ANALYSIS
# For a multiloop circuit with two voltage sources (E1 and E2) and three resistors (R1, R2, R3)
# We need to apply Kirchhoff's Laws to set up the governing equations

print("=== MULTILOOP CIRCUIT CONFIGURATION ===")
print(f"Circuit Components:")
print(f"  R1 = {R1} Ω")
print(f"  R2 = {R2} Ω") 
print(f"  R3 = {R3} Ω")
print(f"  E1 = {E1_theor} V")
print(f"  E2 = {E2_theor} V")

print(f"\n=== CIRCUIT DESCRIPTION ===")
print("Typical multiloop circuit configuration:")
print("- Two voltage sources (E1 and E2)")
print("- Three resistors (R1, R2, R3)")
print("- Two independent loops")
print("- Three branch currents (i1, i2, i3)")

print(f"\n=== STEP 1: DEFINE CURRENT DIRECTIONS ===")
print("Define current directions (assume clockwise positive for each loop):")
print("  i1 = current through R1")
print("  i2 = current through R2") 
print("  i3 = current through R3")

print(f"\n=== STEP 2: APPLY KIRCHHOFF'S CURRENT LAW (KCL) ===")
print("At the junction node:")
print("  Sum of currents entering = Sum of currents leaving")
print("  i1 = i2 + i3")
print("  This gives us Equation 1: i1 - i2 - i3 = 0")

print(f"\n=== STEP 3: APPLY KIRCHHOFF'S VOLTAGE LAW (KVL) ===")
print("For each independent loop, the sum of voltage drops equals the sum of voltage sources")

print(f"\nLoop 1 (containing E1, R1, and R2):")
print("  Following the loop clockwise:")
print("  E1 - i1*R1 - i2*R2 = 0")
print("  Rearranging: i1*R1 + i2*R2 = E1")
print(f"  Substituting values: {R1}*i1 + {R2}*i2 = {E1_theor}")
print("  This gives us Equation 2: 1000*i1 + 2200*i2 = 6.0")

print(f"\nLoop 2 (containing E2, R2, and R3):")
print("  Following the loop clockwise:")
print("  E2 - i2*R2 - i3*R3 = 0")
print("  Rearranging: i2*R2 + i3*R3 = E2")
print(f"  Substituting values: {R2}*i2 + {R3}*i3 = {E2_theor}")
print("  This gives us Equation 3: 2200*i2 + 3300*i3 = 10.98")

print(f"\n=== SYSTEM OF EQUATIONS ===")
print("The multiloop circuit is governed by this system of three equations:")
print("  Equation 1 (KCL): i1 - i2 - i3 = 0")
print("  Equation 2 (KVL Loop 1): 1000*i1 + 2200*i2 = 6.0")
print("  Equation 3 (KVL Loop 2): 2200*i2 + 3300*i3 = 10.98")

print(f"\n=== MATRIX FORM ===")
print("In matrix form: [A][I] = [B]")
print("Where:")
print("  [A] = coefficient matrix")
print("  [I] = current vector [i1, i2, i3]ᵀ")
print("  [B] = voltage vector")
print()
print("  [1   -1   -1] [i1]   [0    ]")
print("  [1000 2200  0] [i2] = [6.0  ]")
print("  [0   2200 3300] [i3]   [10.98]")

# Create the coefficient matrix and voltage vector for later solution
print(f"\n=== COEFFICIENT MATRIX AND VECTORS ===")
import numpy as np

A = np.array([
    [1, -1, -1],      # KCL equation
    [1000, 2200, 0],  # KVL Loop 1
    [0, 2200, 3300]   # KVL Loop 2
])

B = np.array([0, 6.0, 10.98])  # Right-hand side vector

print("Coefficient matrix A:")
print(A)
print(f"\nVoltage vector B:")
print(B)

print(f"\n=== EQUATIONS SUMMARY ===")
print("These three equations completely describe the multiloop circuit:")
print("1. KCL at junction: i1 = i2 + i3")
print("2. KVL Loop 1: E1 = i1*R1 + i2*R2")
print("3. KVL Loop 2: E2 = i2*R2 + i3*R3")
print("\nSolving this system will give us the three unknown currents i1, i2, and i3.")
```

2. Solve the equations in order to find the theoretical values of $i_{\text{1}}$, $i_{\text{2}}$ and $i_{\text{3}}$

```python
# SOLVE THE SYSTEM OF EQUATIONS FOR MULTILOOP CIRCUIT
# We have the system from the previous cell:
# Equation 1 (KCL): i1 - i2 - i3 = 0
# Equation 2 (KVL Loop 1): 1000*i1 + 2200*i2 = 6.0
# Equation 3 (KVL Loop 2): 2200*i2 + 3300*i3 = 10.98

print("=== SOLVING THE MULTILOOP CIRCUIT EQUATIONS ===")

# Use the coefficient matrix A and vector B from the previous cell
# If they don't exist, recreate them
try:
    print(f"Using existing coefficient matrix A and vector B")
except:
    import numpy as np
    A = np.array([
        [1, -1, -1],      # KCL equation
        [1000, 2200, 0],  # KVL Loop 1
        [0, 2200, 3300]   # KVL Loop 2
    ])
    B = np.array([0, 6.0, 10.98])

print("System of equations in matrix form: [A][I] = [B]")
print("Coefficient matrix A:")
print(A)
print(f"\nRight-hand side vector B:")
print(B)

# Method 1: Using NumPy's linear algebra solver
print(f"\n=== METHOD 1: NUMPY LINEAR SOLVER ===")
currents = np.linalg.solve(A, B)

i1_multiloop = currents[0]
i2_multiloop = currents[1] 
i3_multiloop = currents[2]

print(f"Solution vector [i1, i2, i3]:")
print(f"i1 = {i1_multiloop:.6f} A = {i1_multiloop*1000:.3f} mA")
print(f"i2 = {i2_multiloop:.6f} A = {i2_multiloop*1000:.3f} mA")
print(f"i3 = {i3_multiloop:.6f} A = {i3_multiloop*1000:.3f} mA")

# Method 2: Manual substitution method for verification
print(f"\n=== METHOD 2: SUBSTITUTION METHOD (VERIFICATION) ===")
print("Step 1: From Equation 1 (KCL): i1 = i2 + i3")
print("Step 2: Substitute into Equation 2:")
print("  1000*(i2 + i3) + 2200*i2 = 6.0")
print("  1000*i2 + 1000*i3 + 2200*i2 = 6.0")
print("  3200*i2 + 1000*i3 = 6.0  ... (Equation 2')")

print("\nStep 3: We now have two equations with two unknowns:")
print("  Equation 2': 3200*i2 + 1000*i3 = 6.0")
print("  Equation 3:  2200*i2 + 3300*i3 = 10.98")

print("\nStep 4: Solve using elimination method")
# From Equation 2': i3 = (6.0 - 3200*i2)/1000
# Substitute into Equation 3:
# 2200*i2 + 3300*(6.0 - 3200*i2)/1000 = 10.98
# 2200*i2 + 3.3*(6.0 - 3200*i2) = 10.98
# 2200*i2 + 19.8 - 10560*i2 = 10.98
# -8360*i2 = 10.98 - 19.8 = -8.82
# i2 = 8.82/8360

i2_manual = 8.82 / 8360
i3_manual = (6.0 - 3200 * i2_manual) / 1000
i1_manual = i2_manual + i3_manual

print(f"Manual calculation results:")
print(f"i2 = {i2_manual:.6f} A = {i2_manual*1000:.3f} mA")
print(f"i3 = {i3_manual:.6f} A = {i3_manual*1000:.3f} mA") 
print(f"i1 = {i1_manual:.6f} A = {i1_manual*1000:.3f} mA")

# Verification: Check if both methods give the same results
print(f"\n=== VERIFICATION OF METHODS ===")
print(f"Difference in i1: {abs(i1_multiloop - i1_manual):.8f} A")
print(f"Difference in i2: {abs(i2_multiloop - i2_manual):.8f} A")
print(f"Difference in i3: {abs(i3_multiloop - i3_manual):.8f} A")

if (abs(i1_multiloop - i1_manual) < 1e-6 and 
    abs(i2_multiloop - i2_manual) < 1e-6 and 
    abs(i3_multiloop - i3_manual) < 1e-6):
    print("✓ Both methods give identical results")
else:
    print("✗ Methods disagree - check calculations")

# Check that the solution satisfies all original equations
print(f"\n=== EQUATION VERIFICATION ===")
eq1_check = i1_multiloop - i2_multiloop - i3_multiloop
eq2_check = 1000*i1_multiloop + 2200*i2_multiloop
eq3_check = 2200*i2_multiloop + 3300*i3_multiloop

print(f"Equation 1 (should = 0): i1 - i2 - i3 = {eq1_check:.8f}")
print(f"Equation 2 (should = 6.0): 1000*i1 + 2200*i2 = {eq2_check:.6f}")
print(f"Equation 3 (should = 10.98): 2200*i2 + 3300*i3 = {eq3_check:.6f}")

print(f"\n=== THEORETICAL CURRENT VALUES FOR MULTILOOP CIRCUIT ===")
print(f"{'Current':<10} {'Value (A)':<12} {'Value (mA)':<12} {'Direction':<15}")
print("-" * 55)
print(f"{'i1':<10} {i1_multiloop:<12.6f} {i1_multiloop*1000:<12.3f} {'Through R1':<15}")
print(f"{'i2':<10} {i2_multiloop:<12.6f} {i2_multiloop*1000:<12.3f} {'Through R2':<15}")
print(f"{'i3':<10} {i3_multiloop:<12.6f} {i3_multiloop*1000:<12.3f} {'Through R3':<15}")

# Store values for later use
print(f"\n=== STORED VALUES FOR MULTILOOP CIRCUIT ===")
print(f"i1_multiloop_theoretical = {i1_multiloop:.6f} A")
print(f"i2_multiloop_theoretical = {i2_multiloop:.6f} A")
print(f"i3_multiloop_theoretical = {i3_multiloop:.6f} A")

# Physical interpretation
print(f"\n=== PHYSICAL INTERPRETATION ===")
print(f"Current directions and magnitudes:")
if i1_multiloop > 0:
    print(f"- i1 = {i1_multiloop*1000:.3f} mA flows in assumed positive direction through R1")
else:
    print(f"- i1 = {abs(i1_multiloop)*1000:.3f} mA flows opposite to assumed direction through R1")
    
if i2_multiloop > 0:
    print(f"- i2 = {i2_multiloop*1000:.3f} mA flows in assumed positive direction through R2")
else:
    print(f"- i2 = {abs(i2_multiloop)*1000:.3f} mA flows opposite to assumed direction through R2")
    
if i3_multiloop > 0:
    print(f"- i3 = {i3_multiloop*1000:.3f} mA flows in assumed positive direction through R3")
else:
    print(f"- i3 = {abs(i3_multiloop)*1000:.3f} mA flows opposite to assumed direction through R3")

print(f"\nKirchhoff's Current Law verification:")
print(f"At junction: i1 = i2 + i3")
print(f"{i1_multiloop*1000:.3f} mA = {i2_multiloop*1000:.3f} mA + {i3_multiloop*1000:.3f} mA")
print(f"{i1_multiloop*1000:.3f} mA = {(i2_multiloop + i3_multiloop)*1000:.3f} mA ✓")
```

```python
# VERIFY KIRCHHOFF'S CURRENT LAW (KCL) FOR MULTILOOP CIRCUIT
# KCL states that the sum of currents entering a node equals the sum of currents leaving the node
# For our multiloop circuit: At the junction node, i1 = i2 + i3

print("=== KIRCHHOFF'S CURRENT LAW VERIFICATION ===")
print("KCL Statement: The algebraic sum of currents at any junction is zero")
print("Alternative form: Current entering = Current leaving")
print()

# Verify KCL using theoretical values
print("=== THEORETICAL VALUES VERIFICATION ===")
print(f"Theoretical current values:")
print(f"  i1 = {i1_multiloop:.6f} A = {i1_multiloop*1000:.3f} mA")
print(f"  i2 = {i2_multiloop:.6f} A = {i2_multiloop*1000:.3f} mA")
print(f"  i3 = {i3_multiloop:.6f} A = {i3_multiloop*1000:.3f} mA")

print(f"\nAt the main junction node:")
print(f"Current entering: i1 = {i1_multiloop*1000:.3f} mA")
print(f"Current leaving: i2 + i3 = {i2_multiloop*1000:.3f} + {i3_multiloop*1000:.3f} = {(i2_multiloop + i3_multiloop)*1000:.3f} mA")

theoretical_kcl_error = abs(i1_multiloop - (i2_multiloop + i3_multiloop))
print(f"\nKCL Check: i1 - (i2 + i3) = {theoretical_kcl_error:.8f} A")
print(f"Error magnitude: {theoretical_kcl_error*1000:.6f} mA")

if theoretical_kcl_error < 1e-6:
    print("✓ KCL satisfied for theoretical values (within numerical precision)")
else:
    print(f"✗ KCL not satisfied for theoretical values - error: {theoretical_kcl_error:.8f} A")

# Verify KCL using experimental values
print(f"\n=== EXPERIMENTAL VALUES VERIFICATION ===")
print(f"Experimental current values:")
print(f"  i1_exp = {i1_multiloop_experimental:.6f} A = {i1_multiloop_experimental*1000:.3f} mA")
print(f"  i2_exp = {i2_multiloop_experimental:.6f} A = {i2_multiloop_experimental*1000:.3f} mA")
print(f"  i3_exp = {i3_multiloop_experimental:.6f} A = {i3_multiloop_experimental*1000:.3f} mA")

print(f"\nAt the main junction node:")
print(f"Current entering: i1_exp = {i1_multiloop_experimental*1000:.3f} mA")
print(f"Current leaving: i2_exp + i3_exp = {i2_multiloop_experimental*1000:.3f} + {i3_multiloop_experimental*1000:.3f} = {(i2_multiloop_experimental + i3_multiloop_experimental)*1000:.3f} mA")

experimental_kcl_error = abs(i1_multiloop_experimental - (i2_multiloop_experimental + i3_multiloop_experimental))
print(f"\nKCL Check: i1_exp - (i2_exp + i3_exp) = {experimental_kcl_error:.6f} A")
print(f"Error magnitude: {experimental_kcl_error*1000:.3f} mA")

# Calculate percent error in KCL satisfaction
if i1_multiloop_experimental != 0:
    kcl_percent_error = (experimental_kcl_error / i1_multiloop_experimental) * 100
    print(f"KCL percent error: {kcl_percent_error:.2f}%")
else:
    print("Cannot calculate KCL percent error (i1_exp = 0)")

# Assessment of experimental KCL satisfaction
if experimental_kcl_error < 0.0001:  # 0.1 mA tolerance
    print("✓ KCL reasonably satisfied for experimental values")
elif experimental_kcl_error < 0.0005:  # 0.5 mA tolerance
    print("⚠ KCL approximately satisfied for experimental values (small discrepancy)")
else:
    print("✗ KCL not well satisfied for experimental values (significant discrepancy)")

# Comparison table
print(f"\n=== KCL COMPARISON TABLE ===")
print(f"{'Source':<15} {'i1 (mA)':<10} {'i2 + i3 (mA)':<12} {'Difference (mA)':<15} {'Status':<20}")
print("-" * 75)

theoretical_sum = (i2_multiloop + i3_multiloop) * 1000
experimental_sum = (i2_multiloop_experimental + i3_multiloop_experimental) * 1000

print(f"{'Theoretical':<15} {i1_multiloop*1000:<10.3f} {theoretical_sum:<12.3f} {theoretical_kcl_error*1000:<15.6f} {'Perfect':<20}")
print(f"{'Experimental':<15} {i1_multiloop_experimental*1000:<10.3f} {experimental_sum:<12.3f} {experimental_kcl_error*1000:<15.3f} {'Good' if experimental_kcl_error < 0.0005 else 'Poor':<20}")

# Analysis of current distribution
print(f"\n=== CURRENT DISTRIBUTION ANALYSIS ===")
print(f"Theoretical current distribution:")
if i2_multiloop > 0 and i3_multiloop > 0:
    i2_fraction_theoretical = i2_multiloop / i1_multiloop * 100
    i3_fraction_theoretical = i3_multiloop / i1_multiloop * 100
    print(f"  i2 fraction: {i2_fraction_theoretical:.1f}% of total current")
    print(f"  i3 fraction: {i3_fraction_theoretical:.1f}% of total current")
    print(f"  Total: {i2_fraction_theoretical + i3_fraction_theoretical:.1f}% (should be 100%)")

print(f"\nExperimental current distribution:")
if i2_multiloop_experimental > 0 and i3_multiloop_experimental > 0 and i1_multiloop_experimental > 0:
    i2_fraction_experimental = i2_multiloop_experimental / i1_multiloop_experimental * 100
    i3_fraction_experimental = i3_multiloop_experimental / i1_multiloop_experimental * 100
    print(f"  i2 fraction: {i2_fraction_experimental:.1f}% of total current")
    print(f"  i3 fraction: {i3_fraction_experimental:.1f}% of total current")
    print(f"  Total: {i2_fraction_experimental + i3_fraction_experimental:.1f}% (should be 100%)")

# Physical interpretation
print(f"\n=== PHYSICAL INTERPRETATION ===")
print("Kirchhoff's Current Law verification:")
if theoretical_kcl_error < 1e-6:
    print("✓ Theoretical calculations perfectly satisfy KCL")
else:
    print("✗ Theoretical calculations have computational errors")

if experimental_kcl_error < 0.0005:
    print("✓ Experimental measurements reasonably satisfy KCL")
    print("  This confirms proper circuit setup and measurement technique")
else:
    print("✗ Experimental measurements show significant KCL violations")
    print("  This suggests measurement errors, circuit faults, or instrumentation issues")

# Recommendations
print(f"\n=== RECOMMENDATIONS ===")
if experimental_kcl_error > 0.0005:
    print("Suggested improvements for better KCL satisfaction:")
    print("1. Check all connections for loose contacts")
    print("2. Verify ammeter calibration and proper placement")
    print("3. Ensure circuit is in steady state before measurements")
    print("4. Use meters with higher precision/resolution")
    print("5. Repeat measurements multiple times and average results")

# Store verification results
print(f"\n=== STORED VERIFICATION RESULTS ===")
print(f"theoretical_kcl_error = {theoretical_kcl_error:.8f} A")
print(f"experimental_kcl_error = {experimental_kcl_error:.6f} A")
print(f"kcl_percent_error = {kcl_percent_error:.2f}%" if 'kcl_percent_error' in locals() else "kcl_percent_error = N/A")

# Summary
print(f"\n=== KCL VERIFICATION SUMMARY ===")
print(f"Kirchhoff's Current Law: ∑ I_in = ∑ I_out")
print(f"At junction: i1 = i2 + i3")
print(f"Theoretical verification: {'PASSED' if theoretical_kcl_error < 1e-6 else 'FAILED'}")
print(f"Experimental verification: {'PASSED' if experimental_kcl_error < 0.0005 else 'FAILED'}")
print(f"Overall assessment: {'Current conservation verified' if experimental_kcl_error < 0.0005 else 'Current conservation not well satisfied'}")
```

```python
# VERIFY KIRCHHOFF'S VOLTAGE LAW (KVL) FOR EACH LOOP IN MULTILOOP CIRCUIT
# KVL states that the sum of voltage drops around any closed loop equals the sum of voltage sources
# Loop 1: E1 = V1 + V2 (contains E1, R1, R2)
# Loop 2: E2 = V2 + V3 (contains E2, R2, R3)

print("=== KIRCHHOFF'S VOLTAGE LAW VERIFICATION ===")
print("KVL Statement: The algebraic sum of voltages around any closed loop is zero")
print("Alternative form: Sum of voltage sources = Sum of voltage drops")
print()

# Calculate experimental voltages from experimental currents using Ohm's Law
print("=== EXPERIMENTAL VOLTAGE CALCULATIONS ===")
V1_multiloop_exp_calculated = i1_multiloop_experimental * R1
V2_multiloop_exp_calculated = i2_multiloop_experimental * R2
V3_multiloop_exp_calculated = i3_multiloop_experimental * R3

print(f"Voltages calculated from experimental currents (V = I × R):")
print(f"  V1_exp_calc = {i1_multiloop_experimental:.6f} A × {R1} Ω = {V1_multiloop_exp_calculated:.3f} V")
print(f"  V2_exp_calc = {i2_multiloop_experimental:.6f} A × {R2} Ω = {V2_multiloop_exp_calculated:.3f} V")
print(f"  V3_exp_calc = {i3_multiloop_experimental:.6f} A × {R3} Ω = {V3_multiloop_exp_calculated:.3f} V")

print(f"\nDirect experimental voltage measurements:")
print(f"  V1_exp_measured = {V1_multiloop_experimental:.3f} V")
print(f"  V2_exp_measured = {V2_multiloop_experimental:.3f} V")
print(f"  V3_exp_measured = {V3_multiloop_experimental:.3f} V")

# LOOP 1 VERIFICATION: E1 = V1 + V2
print(f"\n=== LOOP 1 VERIFICATION (E1, R1, R2) ===")
print(f"Loop 1 equation: E1 = V1 + V2")
print(f"Voltage source: E1 = {E1_theor} V")

# Theoretical verification
print(f"\nTheoretical verification:")
theoretical_loop1_sum = V1_multiloop + V2_multiloop
theoretical_loop1_error = abs(E1_theor - theoretical_loop1_sum)
print(f"  V1_theoretical + V2_theoretical = {V1_multiloop:.6f} + {V2_multiloop:.6f} = {theoretical_loop1_sum:.6f} V")
print(f"  KVL check: E1 - (V1 + V2) = {E1_theor} - {theoretical_loop1_sum:.6f} = {theoretical_loop1_error:.6f} V")

if theoretical_loop1_error < 0.001:
    print("  ✓ Loop 1 KVL satisfied for theoretical values")
else:
    print(f"  ✗ Loop 1 KVL not satisfied for theoretical values - error: {theoretical_loop1_error:.6f} V")

# Experimental verification using calculated voltages
print(f"\nExperimental verification (using calculated voltages):")
experimental_loop1_sum_calc = V1_multiloop_exp_calculated + V2_multiloop_exp_calculated
experimental_loop1_error_calc = abs(E1_theor - experimental_loop1_sum_calc)
print(f"  V1_exp_calc + V2_exp_calc = {V1_multiloop_exp_calculated:.3f} + {V2_multiloop_exp_calculated:.3f} = {experimental_loop1_sum_calc:.3f} V")
print(f"  KVL check: E1 - (V1 + V2) = {E1_theor} - {experimental_loop1_sum_calc:.3f} = {experimental_loop1_error_calc:.3f} V")

# Experimental verification using measured voltages
print(f"\nExperimental verification (using measured voltages):")
experimental_loop1_sum_measured = V1_multiloop_experimental + V2_multiloop_experimental
experimental_loop1_error_measured = abs(E1_theor - experimental_loop1_sum_measured)
print(f"  V1_exp_meas + V2_exp_meas = {V1_multiloop_experimental:.3f} + {V2_multiloop_experimental:.3f} = {experimental_loop1_sum_measured:.3f} V")
print(f"  KVL check: E1 - (V1 + V2) = {E1_theor} - {experimental_loop1_sum_measured:.3f} = {experimental_loop1_error_measured:.3f} V")

# Calculate percent errors for Loop 1
loop1_percent_error_calc = (experimental_loop1_error_calc / E1_theor) * 100
loop1_percent_error_measured = (experimental_loop1_error_measured / E1_theor) * 100

print(f"  Loop 1 percent error (calculated): {loop1_percent_error_calc:.2f}%")
print(f"  Loop 1 percent error (measured): {loop1_percent_error_measured:.2f}%")

# LOOP 2 VERIFICATION: E2 = V2 + V3
print(f"\n=== LOOP 2 VERIFICATION (E2, R2, R3) ===")
print(f"Loop 2 equation: E2 = V2 + V3")
print(f"Voltage source: E2 = {E2_theor} V")

# Theoretical verification
print(f"\nTheoretical verification:")
theoretical_loop2_sum = V2_multiloop + V3_multiloop
theoretical_loop2_error = abs(E2_theor - theoretical_loop2_sum)
print(f"  V2_theoretical + V3_theoretical = {V2_multiloop:.6f} + {V3_multiloop:.6f} = {theoretical_loop2_sum:.6f} V")
print(f"  KVL check: E2 - (V2 + V3) = {E2_theor} - {theoretical_loop2_sum:.6f} = {theoretical_loop2_error:.6f} V")

if theoretical_loop2_error < 0.001:
    print("  ✓ Loop 2 KVL satisfied for theoretical values")
else:
    print(f"  ✗ Loop 2 KVL not satisfied for theoretical values - error: {theoretical_loop2_error:.6f} V")

# Experimental verification using calculated voltages
print(f"\nExperimental verification (using calculated voltages):")
experimental_loop2_sum_calc = V2_multiloop_exp_calculated + V3_multiloop_exp_calculated
experimental_loop2_error_calc = abs(E2_theor - experimental_loop2_sum_calc)
print(f"  V2_exp_calc + V3_exp_calc = {V2_multiloop_exp_calculated:.3f} + {V3_multiloop_exp_calculated:.3f} = {experimental_loop2_sum_calc:.3f} V")
print(f"  KVL check: E2 - (V2 + V3) = {E2_theor} - {experimental_loop2_sum_calc:.3f} = {experimental_loop2_error_calc:.3f} V")

# Experimental verification using measured voltages
print(f"\nExperimental verification (using measured voltages):")
experimental_loop2_sum_measured = V2_multiloop_experimental + V3_multiloop_experimental
experimental_loop2_error_measured = abs(E2_theor - experimental_loop2_sum_measured)
print(f"  V2_exp_meas + V3_exp_meas = {V2_multiloop_experimental:.3f} + {V3_multiloop_experimental:.3f} = {experimental_loop2_sum_measured:.3f} V")
print(f"  KVL check: E2 - (V2 + V3) = {E2_theor} - {experimental_loop2_sum_measured:.3f} = {experimental_loop2_error_measured:.3f} V")

# Calculate percent errors for Loop 2
loop2_percent_error_calc = (experimental_loop2_error_calc / E2_theor) * 100
loop2_percent_error_measured = (experimental_loop2_error_measured / E2_theor) * 100

print(f"  Loop 2 percent error (calculated): {loop2_percent_error_calc:.2f}%")
print(f"  Loop 2 percent error (measured): {loop2_percent_error_measured:.2f}%")

# COMPREHENSIVE COMPARISON TABLE
print(f"\n=== KVL VERIFICATION SUMMARY TABLE ===")
print(f"{'Loop':<8} {'Source (V)':<12} {'Method':<15} {'Sum (V)':<10} {'Error (V)':<12} {'Error %':<10} {'Status':<15}")
print("-" * 90)

print(f"{'Loop 1':<8} {E1_theor:<12.2f} {'Theoretical':<15} {theoretical_loop1_sum:<10.6f} {theoretical_loop1_error:<12.6f} {(theoretical_loop1_error/E1_theor)*100:<10.2f} {'Perfect' if theoretical_loop1_error < 0.001 else 'Poor':<15}")
print(f"{'Loop 1':<8} {E1_theor:<12.2f} {'Exp-Calculated':<15} {experimental_loop1_sum_calc:<10.3f} {experimental_loop1_error_calc:<12.3f} {loop1_percent_error_calc:<10.2f} {'Good' if experimental_loop1_error_calc < 0.5 else 'Poor':<15}")
print(f"{'Loop 1':<8} {E1_theor:<12.2f} {'Exp-Measured':<15} {experimental_loop1_sum_measured:<10.3f} {experimental_loop1_error_measured:<12.3f} {loop1_percent_error_measured:<10.2f} {'Good' if experimental_loop1_error_measured < 0.5 else 'Poor':<15}")

print(f"{'Loop 2':<8} {E2_theor:<12.2f} {'Theoretical':<15} {theoretical_loop2_sum:<10.6f} {theoretical_loop2_error:<12.6f} {(theoretical_loop2_error/E2_theor)*100:<10.2f} {'Perfect' if theoretical_loop2_error < 0.001 else 'Poor':<15}")
print(f"{'Loop 2':<8} {E2_theor:<12.2f} {'Exp-Calculated':<15} {experimental_loop2_sum_calc:<10.3f} {experimental_loop2_error_calc:<12.3f} {loop2_percent_error_calc:<10.2f} {'Good' if experimental_loop2_error_calc < 0.5 else 'Poor':<15}")
print(f"{'Loop 2':<8} {E2_theor:<12.2f} {'Exp-Measured':<15} {experimental_loop2_sum_measured:<10.3f} {experimental_loop2_error_measured:<12.3f} {loop2_percent_error_measured:<10.2f} {'Good' if experimental_loop2_error_measured < 0.5 else 'Poor':<15}")

# OVERALL ASSESSMENT
print(f"\n=== OVERALL KVL ASSESSMENT ===")
avg_theoretical_error = (theoretical_loop1_error + theoretical_loop2_error) / 2
avg_experimental_error_calc = (experimental_loop1_error_calc + experimental_loop2_error_calc) / 2
avg_experimental_error_measured = (experimental_loop1_error_measured + experimental_loop2_error_measured) / 2

print(f"Average errors:")
print(f"  Theoretical: {avg_theoretical_error:.6f} V")
print(f"  Experimental (calculated): {avg_experimental_error_calc:.3f} V")
print(f"  Experimental (measured): {avg_experimental_error_measured:.3f} V")

print(f"\nKVL verification status:")
theoretical_kvl_status = "PASSED" if avg_theoretical_error < 0.001 else "FAILED"
experimental_kvl_status_calc = "PASSED" if avg_experimental_error_calc < 1.0 else "FAILED"
experimental_kvl_status_measured = "PASSED" if avg_experimental_error_measured < 1.0 else "FAILED"

print(f"  Theoretical calculations: {theoretical_kvl_status}")
print(f"  Experimental (calculated voltages): {experimental_kvl_status_calc}")
print(f"  Experimental (measured voltages): {experimental_kvl_status_measured}")

# ANALYSIS OF DISCREPANCIES
print(f"\n=== ANALYSIS OF VOLTAGE MEASUREMENT DISCREPANCIES ===")
V1_measurement_error = abs(V1_multiloop_exp_calculated - V1_multiloop_experimental)
V2_measurement_error = abs(V2_multiloop_exp_calculated - V2_multiloop_experimental)
V3_measurement_error = abs(V3_multiloop_exp_calculated - V3_multiloop_experimental)

print(f"Comparison of calculated vs measured voltages:")
print(f"  V1: Calculated = {V1_multiloop_exp_calculated:.3f} V, Measured = {V1_multiloop_experimental:.3f} V, Difference = {V1_measurement_error:.3f} V")
print(f"  V2: Calculated = {V2_multiloop_exp_calculated:.3f} V, Measured = {V2_multiloop_experimental:.3f} V, Difference = {V2_measurement_error:.3f} V")
print(f"  V3: Calculated = {V3_multiloop_exp_calculated:.3f} V, Measured = {V3_multiloop_experimental:.3f} V, Difference = {V3_measurement_error:.3f} V")

# FINAL SUMMARY
print(f"\n=== FINAL KVL VERIFICATION SUMMARY ===")
print(f"Kirchhoff's Voltage Law: ∑ V_sources = ∑ V_drops")
print(f"Loop 1 (E1 = V1 + V2): Theoretical {theoretical_kvl_status}, Experimental {experimental_kvl_status_measured}")
print(f"Loop 2 (E2 = V2 + V3): Theoretical {theoretical_kvl_status}, Experimental {experimental_kvl_status_measured}")
print(f"Overall KVL verification: {'SATISFACTORY' if theoretical_kvl_status == 'PASSED' else 'NEEDS IMPROVEMENT'}")

if experimental_kvl_status_measured == "FAILED":
    print(f"\nRecommendations for experimental improvement:")
    print(f"1. Verify voltage measurement instrument calibration")
    print(f"2. Check for consistent measurement polarity")
    print(f"3. Ensure stable circuit conditions during measurements")
    print(f"4. Verify all circuit connections are secure")
    print(f"5. Consider instrument loading effects on the circuit")
```
